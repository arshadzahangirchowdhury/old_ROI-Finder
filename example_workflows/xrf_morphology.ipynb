```python
def _load_xrf_tif_files(base_path = '/data02/AZC/annotated_xrf_simple/',
                   class_type = 'accept',
                   exp_tag= 'bnp_fly0001_2018_1',
                   elm_channel = 'Ca'
                   ):
    
    
    path_to_tif= os.path.join(base_path,class_type,exp_tag,elm_channel, '*.tif')
    
    print(path_to_tif)
    
    img_list = []
    
    for img in glob.glob(path_to_tif):
        n= tiff.imread(img)
        img_list.append(n)
               
    return np.array(img_list, dtype=np.float32)
    
```

```python
#Normalize data to grayscale

def normalize8(I):
    
    mn = I.min()
      
    mx = I.max()

    mx -= mn

    I = ((I - mn)/mx) * 255
  
    return I.astype(np.uint8)
#     return I.astype(int)
```
